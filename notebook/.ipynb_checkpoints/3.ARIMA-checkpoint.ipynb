```python
import time
import datetime
principio = datetime.datetime.now()
principio
```

```python
# Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sktime.utils.plotting import plot_series
import seaborn as sns
```

```python
# Read Data
y = pd.read_csv('../data/datos semanales.csv')
y.set_index('Fecha', inplace=True)
y
```

```python
#Transformación logarítimica
from sktime.transformations.series.boxcox import LogTransformer
transformer = LogTransformer()
y.index = pd.to_datetime(y.index)
log_y = transformer.fit_transform(y)
```

```python
# Plot Log Data
ax = y.plot(marker='o', linestyle='-')
ax.set_ylabel('Billetes')
ax.set_title('Billetes')
```

```python
# Plot Log Data
ax = log_y.plot(marker='o', linestyle='-')
ax.set_ylabel('Billetes (log)')
ax.set_title('Billetes: Transformación LOG')
```

 # Autocorrelación

 <div class="hr2"> </div>
 
 ## Calculamos la Autocorrelación de la **Serie en Logs**

```python
# Autocorrelation Fuction Package
from sktime.utils.plotting import plot_correlations
```

```python
# Autocorrelation Fuction Original Time Series
plot_correlations(log_y)
```

Con la PACF podemos ver que las dos últimas observaciones son las más importantes y que tienen relevancia de manera positiva con respecto a la semana que ahora vamos a predecir. Es decir, si las dos semanas anteriores, la venta de billetes fue alta, es muy probable que esta también lo sea. Realmente, las nueve últimas semanas determinan la predicción de cada observación siguiente.

```python
# Difference Fuction Package
from sktime.transformations.series.difference import Differencer
```

```python
transf_diff=Differencer(lags=[1])
plot_correlations(transf_diff.fit_transform(log_y))
```

```python
#Manu: esta es la tasa de variación anual (porque lag es 52)
transf_diff=Differencer(lags=[52])
plot_correlations(transf_diff.fit_transform(log_y))
```

```python
# Autocorrelation Fuction
# d=1; D=1; S=4
# 
transf_diff=Differencer(lags=[1, 52])
plot_correlations(transf_diff.fit_transform(log_y))
```

 # Modelos ARIMA

 <div class="redcuadro">

 Vamos a separar la muestra en la parte de estiamción (Training) y la parte de predicción/Verificación (Testing). Quitamos 8 trimestres. <br><br>

```python
# Sktime fucntions
# Forecast horizon and Split function 
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.model_selection import temporal_train_test_split
```

```python
y = y.astype('float64').to_period('W-Sat')
```

```python
# Select Train & test sample
y_train, y_test = temporal_train_test_split(y, test_size=5)
plot_series(y_train, y_test, labels=["Train", "Test"])
# Time Series Size
print(y_train.shape[0], y_test.shape[0])
```

```python
# Forecast Horizon
fh = np.arange(len(y_test)) + 1  # forecasting horizon
fh
```

```python
# Sktime Auto ARIMA Function
from sktime.forecasting.arima import AutoARIMA
# !pip install pmdarima
```

En un notebook complementario hice una búsqueda con un bucle for en el que se comprobase cuáles son los parámetros que minimizan el MAPE. La prueba la hice entre todos los posibles seasonal (que es buleano), stationary (que también es buleano), information_criterion y scoring. La mejor combinación, tras analizar todas las posibles combinaciones, es la que tenemos en la siguiente casilla.

```python
forecaster = AutoARIMA(
        seasonal=True,
        stationary=False,
        information_criterion='aic',
        scoring='mse',
        sp=52,
        suppress_warnings=True)
```

```python
forecaster.fit(y_train)
```

```python
#  Auto ARIMA Model Summary
print(forecaster.summary())
```

```python
fh
```

```python
#  Auto ARIMA Forecast
y_pred = forecaster.predict(fh)
```

```python
# Sktime fucntions
# Forecast Accuracy MAPE & MSE & RMSE
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
```

```python
# Forecast Accuracy
# MAPE 
# Orignal Time Serie

mape = mean_absolute_percentage_error(y_test, y_pred)
mape
```

```python
# Forecast Accuracy
#Mean Squared Error
mse = MeanSquaredError(square_root=True)
mse(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

## Comprobaciones del error

```python
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
from sklearn.metrics import r2_score

mape = mean_absolute_percentage_error(y_test, y_pred)
mse = MeanSquaredError()
mse = mse(y_test, y_pred)
r2 = r2_score (y_test, y_pred)

print(f'MAPE: {mape}')
print(f'MSE: {mse}')
print(f'R2 Score: {r2}')
```

Para medir la precisión de este modelo comprobamos el MAPE, el MSE y el R2 score.

Para elegir modelo nos hemos centrado en el MAPE, que es el que proporciona una medida relativa de la precisión de un modelo de pronóstico al evaluar el porcentaje promedio de error entre las predicciones y los valores reales.

El MSE, que es la suma de los cuadrados de los errores dividido entre el número total de observaciones, penaliza fuertemente los errores de mayor calado.

Con R2 conseguimos saber cuánto de bien explica el modelo la variabilidad total de los datos. Su interpretación varía entre 0 y 1, y un valor más alto indica un mejor ajuste del modelo. A pesar de que el MAPE sea menor aquí que en la predicción diaria, la explicabilidad es ligeramente menor (0,5927 en la predicción diaria y 0,5816 en la semanal).

Esta predicción tiene menos error que la diaria porque el horizonte temporal de predicción es menor: solo hay que predecir 5 valores y no 31.

# Prediccion

```python
# Forecast Horizon
fh = np.arange(5) + 1  # forecasting horizon
fh
```

```python
forecaster = AutoARIMA(
        seasonal=True,
        stationary=False,
        information_criterion='aic',
        scoring='mse',
        sp=52,
        suppress_warnings=True)
```

```python
forecaster.fit(y)
```

```python
print(forecaster.summary())
```

```python
y_pred = forecaster.predict(fh)
```

```python
plot_series(y, y_pred, labels=["Billetes", "Billetes pred"])
```

```python
plot_series(y["2016":], y_pred, labels=["Billetes", "Billetes pred"])
```

```python
bus_pred_ints = forecaster.predict_interval(fh, coverage=0.9)
```

```python
fig, ax = plot_series(y_test, y_pred, labels=["Billetes", "Billetes pred"])
ax.fill_between(
    ax.get_lines()[-1].get_xdata(),
    bus_pred_ints[('Coverage', 0.9, 'lower')],
    bus_pred_ints[('Coverage', 0.9, 'upper')],
    alpha=0.2,
    color=ax.get_lines()[-1].get_c(),
    label=f"90% prediction intervals",
)
ax.legend(loc='upper left')
```

Esta es la predicción para las cinco primeras semanas de 2017
